```python
import google.generativeai as genai
import datetime as date

# Choose a Gemini model.
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

prompt = "You are an automatic parking gate that receive picture and recognizing plate number, recognize plate number input from picture uploaded with a return of JSON. Here sample format of JSON: {‘plat_no’: ‘B 1234 ABC’,‘plat_expired_date’: ‘05-25’,‘vehicle’: ‘car’,‘vehicle_type’: ‘sedan’,‘color’: ‘red’,‘gate_open’: '2024-12-07 12:07:33',‘gate_closed’: ‘N/A’}"
```

```python
# Create a prompt.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('Mazda3.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()

sample_file_1 = PIL.Image.open('Mazda3.jpg')
response = model.generate_content([sample_file_1, prompt])

Markdown(response.text)
```
